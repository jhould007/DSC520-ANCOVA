```python
# Import general libraries
import pandas as pd
import numpy as np

# Import modeling libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.formula.api as smf

# Import visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Import diagnostic libraries
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.anova import AnovaRM

# Examine dataset
df = pd.read_csv("Database MiniPONS.csv", sep=";")
df.head()
```

```python
# Make all columns lowercase for ease of typing
df.columns = df.columns.str.lower()
print("New column names:", df.columns)
print()

# Set "group" values to lowercase
df["group"] = df["group"].str.lower()

# Set "type" values to "typeI", "typeII", "control" and "ud"
df["type"] = np.where(df["type"] == "BD I", "typeI", df["type"])
df["type"] = np.where(df["type"] == "BD II", "typeII", df["type"])
df["type"] = np.where(df["type"] == "Control", "control", df["type"])
df["type"] = np.where(df["type"] == "UD", "ud", df["type"])
print("Unique values of 'type':", df["type"].unique())
```

ANCOVA can be performed when each of the within-group trends have the same slope, are parallel to one another, and are parallel to the overall trend. Make this determination with respect to your data (i.e., the MiniPONS dataset) and state the results.

With this data, we assume that `age` is the covariate because it is often used as such in psychological studies to control for potential age-related effects when comparing groups. We can use an interaction test between the categorical variable `group` and the covariate `age` to determine if the regression slopes are homogeneous. 

<b>Interaction test hypotheses</b><br>
* $H_0$: The regression slopes are equal across groups. 
* $H_a$: The regression slopes differ by group. 
* $\alpha = 0.05.$

```python
# Fit model with interaction term
model = ols("right_answers ~ age * group", data=df).fit()

# Perform interaction test
interaction_test = sm.stats.anova_lm(model, typ=2)
interaction_test
```

We see that the interaction term has a P-value of about $0.65$, which is far greater than $\alpha = 0.05.$ Therefore, we fail to reject $H_0$ and conclude that the regression slopes are equal across groups. Thus, ANCOVA is appropriate for the data.

### Task 5
Perform ANCOVA and address the following:

1. Describe the factors. For example: a) A is categorical, B is continuous covariate. b) A and B are categorical; C is continuous covariate.
2. Calculate the total variability of the response variable.
3. Calculate the unexplained residuals and F-ratios for each factor.
4. Calculate the covariate levels. <hr>

```python
# Fit ANCOVA model
model = smf.ols("right_answers ~ C(group) + age + C(type) + audio_prosody + combined_channel + face_video + body_video + positive_valence + negative_valence + dominant + submissive", data=df)
result = model.fit()

# Summarize result
print(result.summary())
```

<b>Describe the factors. For example: a) A is categorical, B is continuous covariate. b) A and B are categorical; C is continuous covariate.</b>

`group`: Categorical factor <br>
`type`: Categorical factor <br>
`age`: Continuous covariate <br>
`right_answers`: Dependent variable <br>
`audio_prosody`, `combined_channel`, `face_video`, `body_video`, `positive_valence`, `negative_valence`, `dominant`, `submissive`: Continuous factors

<b>Calculate the total variability of the response variable.</b> <br>

This can be done by calculating the SST (total sum of squares) for `right_answers`. 

```python
# Find mean number of right answers
right_answers = df["right_answers"]
mean_right_answers = np.mean(right_answers)

# Calculate total sum of squares (SST)
sst = np.sum((right_answers - mean_right_answers)**2)
print(f"SST: {sst}")
```

<b>Calculate the unexplained residuals and F-ratios for each factor.</b>

```python
# Get ANOVA table
anova_table = anova_lm(result)

# Calculate F-ratio and unexplained residuals for 'group'
sse_group = anova_table["sum_sq"]["Residual"]
ssr_group = anova_table["sum_sq"]["C(group)"]
df_group = anova_table["df"]["C(group)"]
mse_group = ssr_group / df_group
mse_residual = sse_group / (len(df) - len(result.params))
f_ratio_group = mse_group / mse_residual
print(f"MSE for C(group): {mse_residual}")
print(f"F-ratio for C(group): {f_ratio_group}")
print()

# Calculate F-ratio and unexplained residuals for 'group'
sse_type = anova_table["sum_sq"]["Residual"]
ssr_type = anova_table["sum_sq"]["C(type)"]
df_type = anova_table["df"]["C(type)"]
mse_type = ssr_type / df_type
mse_residual = sse_type / (len(df) - len(result.params))
f_ratio_type = mse_type / mse_residual
print(f"MSE for C(type): {mse_residual}")
print(f"F-ratio for C(type): {f_ratio_type}")
print()

# Calculate MSE for each remaining factor
remaining_factors = ["age", "audio_prosody", "combined_channel", "face_video", "body_video", "positive_valence", "negative_valence", "dominant", "submissive"]
for factor in remaining_factors: 
    sse_factor = anova_table["sum_sq"]["Residual"]
    ssr_factor = anova_table["sum_sq"][factor]
    df_factor = anova_table["df"][factor]
    mse_factor = ssr_factor / df_factor
    print(f"MSE for {factor}: {mse_factor}")
```

<b>Verify that residuals are normally distributed.</b>

```python
# Calculate residuals
residuals = result.resid

# Create 1x2 grid
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot boxplot
sns.boxplot(x=residuals, ax=axes[0])
axes[0].set_xlabel("Residuals")
axes[0].set_title("Boxplot of Residuals")

# Plot histogram
sns.histplot(residuals, ax=axes[1])
axes[1].set_xlabel("Residuals")
axes[1].set_title("Histogram of Residuals")

plt.tight_layout()
plt.show()
```

From the histogram, we see a somewhat normal distribution of residuals that is complicated by a few negative outliers. This pattern is also shown in the boxplot. However, a close analysis of the residual <i>values</i> on the x-axis of the plots reveals how small these residuals are, so there should not be a problem. 

<b>Residuals are equally varied. Use boxplots and means against variance and scale as needed.</b>

```python
# Plot residuals vs. fitted values
sns.scatterplot(x=result.fittedvalues, y=residuals)
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs. Fitted Values")
plt.axhline(0, color="red", linestyle="--", linewidth=1)
plt.show()
```

The residuals appear to be in several clusters across the plot, which violates the assumption of equal variance. 

<b>Residuals are independent of one another.</b> 

We will use a Durbin-Watson test here. 
* $H_0$: There is no autocorrelation in the residuals (Durbin-Watson stat close to 2). 
* $H_a$: There is autocorrelation present in the residuals (Durbin-Watson stat far from 2). 
* $\alpha = 0.05.$

```python
# Check for residual independence with Durbin-Watson test
durbin_watson_stat = sm.stats.durbin_watson(residuals)
print(f"Durbin-Watson Statistic: {durbin_watson_stat}")
```

We see that the Durbin-Watson statistic is significantly less than 2, so we conclude that there is significant positive autocorrelation present in the residuals.
